```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import json

def load_court_data(court, case_type, weeks):
    all_cases = []
    for week in weeks:
        file_path = f'retsmder/{court}/{case_type}/{case_type}_uge_{week}.json'
        with open(file_path, 'r', encoding='utf-8') as f:
            cases = json.load(f)
            all_cases.extend(cases)
    return pd.DataFrame(all_cases)

def plot_case_durations(court, case_type, weeks):
    df = load_court_data(court, case_type, weeks)
    
    plt.figure(figsize=(10, 6))
    sns.histplot(data=df, x='Varighed', bins=20)
    plt.title(f'Distribution of Case Durations - {court} {case_type}')
    plt.xlabel('Duration (minutes)')
    plt.ylabel('Count')
    plt.show()

plot_case_durations('Kbenhavn', 'Straffe', [7, 8])
plot_case_durations('Kbenhavn', 'Civile', [7, 8])
plot_case_durations('Hillerd', 'Straffe', [7, 8])
plot_case_durations('Hillerd', 'Civile', [7, 8])
```
